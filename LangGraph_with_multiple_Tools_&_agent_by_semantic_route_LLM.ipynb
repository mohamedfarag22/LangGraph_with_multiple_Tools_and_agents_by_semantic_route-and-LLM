```python
def semantic_layer(state: AgentState) -> AgentState:
    query = state['messages'][0].content  # Extract the query content
    route = rl(query)
    if route.name == "get_time":
        state["action"] = 'get_time'
        return state
        #get_time(state)
    elif route.name == "get_weather":
        state["action"] = "Weather"
        return state
        #Weather(state)
    elif route.name == "search":
        state["action"] = "search"
        return state
        #search(state)
    else:
        state["messages"].append(HumanMessage(content="Sorry, I don't understand your query."))
        state["action"] = None
        return state
```

# Tools

```python

from datetime import datetime
from langchain_core.messages import FunctionMessage
def get_time(state: AgentState) -> AgentState:
    now = datetime.now()
    user_input = state['messages'][0].content
    complete_query = f'''
help the user with provided information to get suitable response:
query:
when should I go to the gym?
information:
The current time is "9:30".
Response:
It's currently "9:30". The best time to go to the gym depends on your schedule and energy levels. Morning workouts can boost your energy for the day.
query:
{user_input}, the current time is {now.strftime("%H:%M")}.
Response: '''
    response = dolly_pipeline(complete_query)
    result = response[0]['generated_text'].strip()
    state["messages"].append(HumanMessage(content=result))
    state['action'] = None
    final = state["messages"][-1].content
    return {"messages": [FunctionMessage(content=str(final), name='get_time')]}

def Weather(state: AgentState) -> AgentState:
    user_input = state['messages'][0].content
    complete_query = f'''
Extract only the city name from the following query:
query:
What's the temperature in Egypt?
Response:
Egypt
query:
{user_input}
Response: '''
    response = dolly_pipeline(complete_query)
    city_name = response[0]['generated_text'].strip()
    state['messages'].append(HumanMessage(content=city_name))

    weather = OpenWeatherMapAPIWrapper()
    weather_data = weather.run(city_name)
    state['messages'].append(HumanMessage(content=weather_data))

    agent2_query = f"Provide concise information based on the user query and the available information. User query: {user_input} Available information: {weather_data}"
    response = dolly_pipeline(agent2_query)
    result = response[0]['generated_text'].strip()
    state["messages"].append(HumanMessage(content=result))
    state['action'] = None
    final = state["messages"][-1].content
    return {"messages": [FunctionMessage(content=str(final), name='weather')]}

def search(state: AgentState) -> AgentState:
    query = state['messages'][0].content
    tool = TavilySearchResults()
    content = tool.invoke({"query": f"{query}"})
    clear_content = [content[c]['content'] for c in range(3)]

    complete_query = f'''
provide Summarized Answer by question user provided with relevant content:
query:
What about LangGraph?
content:
LangGraph is a library for building stateful, multi-actor applications with LLMs, built on top of (and intended to be used with) LangChain.js. It extends the LangChain Expression Language with the ability to coordinate multiple chains (or actors) across multiple steps of computation in a cyclic manner. It is inspired by Pregel and Apache Beam.
Response:
LangGraph is a library that enables the creation of stateful, multi-actor applications using LLMs. It extends LangChain.js by facilitating the coordination of multiple chains or actors through cyclic computation steps, leveraging concepts from Pregel and Apache Beam.
query:
{query}
content:
{clear_content}
Response: '''
    response = dolly_pipeline(complete_query)
    result = response[0]['generated_text'].strip()
    state["messages"].append(HumanMessage(content=result))
    final = state["messages"][-1].content
    return {"messages": [FunctionMessage(content=str(final), name='search')]}
```

# Formating Messages

```python
from typing import TypedDict, Annotated, Sequence
from operator import add
from langchain_core.messages import BaseMessage
from langchain_core.messages import HumanMessage


class AgentState(TypedDict):
    messages: Annotated[Sequence[HumanMessage], add]
    action: str
```

# Condition Function

```python
def where_to_go(state):
  action = state['action']
  called_tool = False

  if action == "search":
    called_tool = True
    return "search"
  elif action == "Weather":
    called_tool = True
    return "get_weather"
  elif action == "get_time":
    called_tool = True
    return "get_time"
  elif action == None:
    called_tool = False
    return "end"
  else:
    called_tool = False
    return "end"
```

# LangGraph

```python

# Define the graph
from langgraph.graph import Graph
from langgraph.graph import StateGraph, END
workflow = StateGraph(AgentState)

# Add nodes for semantic layer (agent) and tools
workflow.add_node("agent", semantic_layer)
workflow.add_node("tool_Time", get_time)
workflow.add_node("tool_Weather", Weather)
workflow.add_node("tool_Search", search)

# Define edges based on the agent's logic
workflow.add_conditional_edges("agent", where_to_go,{
                                                        "search": "tool_Search",
                                                        "get_weather": "tool_Weather",
                                                        "get_time": "tool_Time",
                                                        "end": END
                                                    }
)


# Set entry point and finish point (assuming semantic_layer as entry point)
workflow.set_entry_point("agent")


# Compile the graph
app = workflow.compile()
```
